## Function to read Skeleton (pandas DataFrame) from Kinect raw extracted data with KinectXEFTools (.txt file)

```python
def read_Skeleton(path):
    f=open(path) 
    head=f.readline()
    df=pd.DataFrame()
    
    slice_= lambda A, n=2: [A[i:i+n] for i in range(0, len(A), n)]
    
    c=0
    for line in f:
        x=line.split(',')
        
        l=[]
        ln=[]
        for n,i in enumerate(x):
            if n>8:
                try:
                    int(i)
                    if int(i)==0 and int(x[n+1])==0:
                        pass
                    else:
                        l.append(i)
                        ln.append(n)
                except:pass
        ln_slice=slice_(ln)
        ln_fit=[]
        for n,i in enumerate(ln_slice):
            ln_fit.append([i[0]-n,i[0]-n+2])

        for i in ln_fit:
            x[i[0]:i[1]]=['.'.join(x[i[0]:i[1]])]
            
        ind=[]
        var=[]
        for n,i in enumerate(head.split(',')):
            if 'Rotation' in i:
                pass
            else:
                var.append(i)
                ind.append(n)
        x2=[]
        for n,i in enumerate(x):
            if n in ind:
                x2.append(i)
            else:pass
        
        df[str(c)]=x2
        c+=1
        
    df['Event']=var
    
    return(df)
```

The read_Skeleton() function generates a dataframe with general information for each frame:
['EventIndex',
 'Time',
 'SkeletonId',
 'HandLeftConfidence',
 'HandLeftState',
 'HandRightConfidence',
 'HandRightState']
 and then this fields for each joint in each frame:  ['Joint',
 'Status',
 'PositionX',
 'PositionY',
 'PositionZ']

## Reading .dat file with depth data from KinectXEFTools

```python
def read_Depth(path_depth):
    #DEPTH IMAGES HAVE RESOLUTION 424X512
    with open(path_depth, mode='rb') as file: # b is important -> binary
        fileContent = file.read()
    x=np.fromstring(fileContent[20:],dtype=np.uint16) #first 20 digits have other information

    #Each matriz of 424x512=217088 has 4 digits of information (event number, etc...) in front of it, which must be skipped. 
    #There are 1027 matrices
    
    c=0
    arrays_list=[]
    for i in range(0,2058,2):
        try:
            x1=x[4+108546*i:108546*(i+2)]
            #if you want to save the depth images 
            #cv2.imwrite('C:/path_to_directory/depth_array_{}.jpeg'.format(c),x1.reshape(424,512))
            arrays_list.append(x1.reshape(424,512))
            c+=1
        except:pass
    return(arrays_list)
```

The read_Depth() funtion generates a list of arrays, each one is a depth image that contains the depth of each pixel

```python
#EXAMPLE
path_depth='C:/path_to_dat_file/record.dat'
arrays=read_Depth(path_depth)
```
